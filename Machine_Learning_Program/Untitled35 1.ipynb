```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting the data
plt.plot(x, y)

# Labeling the x and y axes
plt.xlabel("X axis")
plt.ylabel("Y axis")

# Adding a title to the plot
plt.title("Line Plot")

# Showing the plot
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D', 'E']
data = [2, 4, 6, 8, 10]

# Plotting the data
plt.bar(labels, data)

# Labeling the y axis
plt.ylabel("Y axis")

# Adding a title to the plot
plt.title("Bar Plot")

# Showing the plot
plt.show()
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting the data
plt.scatter(x, y)

# Labeling the x and y axes
plt.xlabel("X axis")
plt.ylabel("Y axis")

# Adding a title to the plot
plt.title("Scatter Plot")

# Showing the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [2, 4, 6, 8, 10, 2, 4, 6, 8, 10, 2, 4, 6, 8, 10]

# Plotting the data
plt.hist(data, bins=4)

# Labeling the x axis
plt.xlabel("X axis")

# Adding a title to the plot
plt.title("Histogram")

# Showing the plot
plt.show()
```
